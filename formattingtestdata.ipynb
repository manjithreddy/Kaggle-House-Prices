```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('test.csv')
```

```python
obj = df.isnull().sum()
for key,value in obj.iteritems():
    print(key,",",value)
```

```python
df.drop(columns=['Alley','FireplaceQu','PoolQC','Fence','MiscFeature','Id'],inplace=True)
```

```python
df.shape
```

```python
df.info()
```

```python
meanfeatures=['LotFrontage','MasVnrArea','BsmtFinSF1','BsmtFinSF2','BsmtUnfSF','TotalBsmtSF','BsmtFullBath','BsmtHalfBath','GarageYrBlt','GarageCars','GarageArea']
for i in meanfeatures:
    df[i]=df[i].fillna(df[i].mean().round(1))
```

```python
modefeatures=['MSZoning','Utilities','Exterior1st','Exterior2nd','MasVnrType','BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','KitchenQual','Functional','GarageType','GarageFinish','GarageQual','GarageCond','SaleType']
for i in modefeatures:
    df[i]=df[i].fillna(df[i].mode()[0])
```

```python
obj = df.isnull().sum()
for key,value in obj.iteritems():
    print(key,",",value)
```
